```python
from sympy import *
from sympy.plotting.pygletplot import PygletPlot as Plot
from IPython.display import display
import math
import matplotlib.pyplot as plt
```

# **#1**

```python
x = symbols('x')
f1 = abs(x-3)*sqrt(x)
f2 = asin(2*x/(1+x**2))
f3 = exp(x)/abs(1+x)
p = plot(f1, f2, f3, (x, -10, 10), show=False, ylim=(-10, 10))
p[0].line_color = 'r'
p[1].line_color = 'g'
p[2].line_color = 'b'
p.show()
```

# **#2**

```python
x, y = symbols('x y')
f1 = log(x)
f2 = Eq(abs(x) + abs(y), 1)
display(f2)
p1 = plot(f1, (x, -2, 2), show=False)
p2 = plot_implicit(f2, show=False)
p1.extend(p2)
p1.show()
```
